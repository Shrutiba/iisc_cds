```python
# First DataFrame
df1 = pd.DataFrame({'id': ['A01', 'A02', 'A03', 'A04'],
                    'Name': ['ABC', 'PQR', 'DEF', 'GHI']})

# Second DataFrame
df2 = pd.DataFrame({'id': ['B05', 'B06', 'B07', 'B08'],
                    'Name': ['XYZ', 'TUV', 'MNO', 'JKL']})

dflist = [df1,df2]
print(dflist)
df1 = pd.concat(dflist)
df1
```

```python
df1 = pd.DataFrame({'id': ['A01', 'A02', 'A03', 'A04'],
                    'Name': ['Ramesh', 'Mohan', 'Shilpa', 'Deepthi']})

df2 = pd.DataFrame({'City': ['MUMBAI', 'PUNE', 'MUMBAI','DELHI'],
                    'Age': ['12', '13', '14', '12']})

df3 = pd.concat([df1,df2], axis=1, join='inner')
print(df3)

df4 = pd.concat([df1,df2], axis=0, join='outer')
print(df4)

df5 = pd.concat([df1,df2])
print(df5)

df6 = pd.concat([df1,df2], axis=1, join='outer')
print(df6)
```
